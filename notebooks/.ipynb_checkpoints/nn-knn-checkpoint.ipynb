```python
import os
import sys
import numpy as np
sys.path.insert(0, os.path.abspath('..'))
sys.path.insert(0, os.path.abspath('../algorithms'))
sys.path.insert(0, os.path.abspath('../scoring'))

from scores import cross_validation_score
from algorithms import nearestneighbor
from nearestneighbor import KnnClassifier
from sklearn.model_selection import train_test_split
from scores import confusion_matrix
import matplotlib.pyplot as plt
```

# k-nearest neighbor algorithm for uci mushroom data-set

```python
mushroom_uci_data = np.genfromtxt('../pre-processed-uci-mushroom-dataset/standardized_mushroom_data.csv', delimiter=",")
mushroom_uci_target = np.genfromtxt('../pre-processed-uci-mushroom-dataset/standardized_mushroom_target.csv', delimiter=",")
```

## Split the uci data into testing and training data 

```python
X_train, X_test, y_train, y_test = train_test_split(mushroom_uci_data, mushroom_uci_target, random_state=1000, train_size=750, test_size=250)
```

## Performance on test set is 98.4% accurate

```python
import matplotlib.pyplot as plt

knn = KnnClassifier(3)
knn.fit(X_train, y_train)
knn.score(X_test, y_test)
```

## Generates confusion matrix

```python
cfm = confusion_matrix()
print(cfm.generate(knn.predict(X_test), y_test))
```

## Creates ROC curve for uci-mushroom repository

```python
#Disclamer used sci-kit learn to generate roc curve as my implementation of knn does not have a pred_proba() function
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
from sklearn.neighbors import KNeighborsClassifier

sci_nn = KNeighborsClassifier(n_neighbors=3)
sci_nn.fit(X_train, y_train)
y_scores = sci_nn.predict_proba(X_test)
sensitivity, true_positive_rate, th = roc_curve(y_test, y_scores[:, 1])
AUC = auc(sensitivity, true_positive_rate)
print("AUC =", AUC)

plt.plot(sensitivity, true_positive_rate, 'r')
plt.plot([0, 1], [0, 1],'g--')
plt.xlim([0, 1])
plt.ylabel('Sensitivity')
plt.ylim([0, 1])
plt.xlabel('1 - Specificity')
plt.title('ROC Curve for knn on UCI-mushroom data-set')
plt.show()
```

# k-nearest neighbor algorithm for image mushroom data-set

```python
mushroom_image_data = np.genfromtxt('../mushroom_images/mushroom_image_data.csv', delimiter=",")
mushroom_image_target = np.genfromtxt('../mushroom_images/mushroom_image_target.csv', delimiter=",")
```

## Split the image data into testing and training data 

```python
X_train, X_test, y_train, y_test = train_test_split(mushroom_image_data, mushroom_image_target, random_state=1000)
```

## Standardising image data

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(X_train)
X_train = ss.transform(X_train)
X_test = ss.transform(X_test)
```

## The top image of the generalisation curve, looks  to be k=17 therefore this will be used for the size of k.

```python
import matplotlib.pyplot as plt

scores = list()
crossvalscores = list()
top_k = 41
for k in range(1,top_k, 1):
    knn = KnnClassifier(k)
    knn.fit(X_train, y_train)
    knnScore = knn.score(X_train, y_train)
    scores.append(knnScore)
    cvs = cross_validation_score()
    crossvalscores.append(np.mean(cvs.score(knn, X_train, y_train, cv=5)))
plt.plot(np.arange(1,top_k, 1), scores)
plt.plot(np.arange(1,top_k, 1), crossvalscores)
plt.xlim(top_k , 0)
plt.xlabel("Model complexity (k)")
plt.ylabel("Accuracy score")
plt.legend(["Train score", "Five fold cross validation score"])
plt.axvline(17, 0, 1, label='k=17', c='red')
```

## Implementing nearest neighbor algorithm on test set

```python
nn = KnnClassifier(17)
nn.fit(X_train, y_train)
nn.score(X_test, y_test)
```

## Generates confusion matrix

```python
cfm = confusion_matrix()
nn = KnnClassifier(17)
nn.fit(X_train, y_train)
print(cfm.generate(nn.predict(X_test), y_test))
```

## Creates ROC curve for uci-mushroom repository

```python
#Disclamer used sci-kit learn to generate roc curve as my implementation does not have a pred_proba() function
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
from sklearn.neighbors import KNeighborsClassifier

sci_nn = KNeighborsClassifier(n_neighbors=17)
sci_nn.fit(X_train, y_train)
y_scores = sci_nn.predict_proba(X_test)
sensitivity, true_positive_rate, th = roc_curve(y_test, y_scores[:, 1])
AUC = auc(sensitivity, true_positive_rate)
print("AUC =", AUC)

plt.plot(sensitivity, true_positive_rate, 'r')
plt.plot([0, 1], [0, 1],'g--')
plt.xlim([0, 1])
plt.ylabel('Sensitivity')
plt.ylim([0, 1])
plt.xlabel('1 - Specificity')
plt.title('ROC Curve for knn on image data-set')
plt.show()
```
